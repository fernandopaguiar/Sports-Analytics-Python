```python
Position = ['Outside Back', 'Center Forward', 'Central Midfielder', 'Left Wing', 'Right Wing', 'Goalkeeper']
```

```python
dados.loc[dados["Name"].str.contains("Bil"), "Position"] = "Outside Back"
dados.loc[dados["Name"].str.contains("Jack"), "Position"] = "Right Wing"
dados.loc[dados["Name"].str.contains("Franco"), "Position"] = "Center Forward"
dados.loc[dados["Name"].str.contains("Elijiah"), "Position"] = "Outside Back"
dados.loc[dados["Name"].str.contains("Micah"), "Position"] = "Central Midfielder"
dados.loc[dados["Name"].str.contains("Aaron"), "Position"] = "Left Wing"
dados.loc[dados["Name"].str.contains("Caden"), "Position"] = "Center Forward"
dados.loc[dados["Name"].str.contains("Trevor"), "Position"] = "Goalkeeper"
dados.loc[dados["Name"].str.contains("Jonathon"), "Position"] = "Outside Back"
dados.loc[dados["Name"].str.contains("Santiago"), "Position"] = "Central Midfielder"
dados.head()
```

```python
dados.describe()
```

```python
# criar matriz de dor
dor = dados['Soreness'].values.reshape(-1, len('dados'))

# criar mapa de calor
fig, ax = plt.subplots(figsize=(15, 8))
ax = sns.heatmap(dor, cmap='coolwarm', center=5, annot=True, fmt='.1f')

ax.set_xticklabels(dados['Date'])

# rotular as linhas com os nomes dos participantes
ax.set_yticklabels(dados['Name'])

# configurar visualizaÃ§Ã£o
plt.title("PontuaÃ§Ãµes de Dor")
plt.xlabel("Data")
plt.ylabel("Participante")

# exibir o mapa de calor
plt.show()
```

```python
max_speed = dados.Max_Speed.mean()
```

```python
# Criar figura e eixos
fig, ax = plt.subplots(figsize=(15, 8))

# Adicionar barras
ax.bar(dados.Position, dados['Max_Speed'], color='blue', label='Barras')

# calcular mÃ©dia da mÃ¡xima velocidade para cada jogador
max_speed_mean = dados.groupby('Position')['Max_Speed'].mean()

# criar lista com a mÃ©dia da mÃ¡xima velocidade para cada jogador, na ordem das posiÃ§Ãµes
max_speed_mean_by_position = max_speed_mean.reindex(dados['Position'])

# adicionar linha com a mÃ©dia da mÃ¡xima velocidade
ax.plot(max_speed_mean_by_position, color='red', label='MÃ©dia')

# Configurar eixos
ax.set_xlabel('PosiÃ§Ã£o')
ax.set_ylabel('Velocidade MÃ¡xima')
ax.set_ylim([0, 40])

# Adicionar legenda
ax.legend()

# Exibir grÃ¡fico
plt.show()
```
